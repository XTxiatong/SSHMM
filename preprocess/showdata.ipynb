```python
import numpy as np    
import json
import math
import matplotlib.pyplot as plt

#北京四环内
region_beijing = json.load(open("region_tencent.json"))
grid_boundary = region_beijing['grid_boundary']['0'] #675个街道的边界
#print grid_boundary.keys() #网格ID

plt.figure()
for r in grid_boundary:
    grid_boundary_r =  np.array(grid_boundary[r])
    plt.plot(grid_boundary_r[:,0],grid_boundary_r[:,1])
plt.show()
```

```python
data = json.load(open("flow_tencent_beijing_April_region.json")) #人流数据, 时间从2018.4.1-4.31， 每半个小时提取一次
print data.keys()
population_beijing = data['population']
inflow_beijing = data['inflow_beijing']
outflow_beijing = data['outflow_beijing']

plt.figure()
for k in population_beijing.keys()[:10]:
    tmp = np.array(population_beijing[k])[:336]
    plt.plot(tmp/np.max(tmp))
plt.show()

plt.figure()
for k in population_beijing.keys()[:10]:
    tmp = np.array(inflow_beijing[k])[:336]
    plt.plot(tmp/np.max(tmp))
plt.show()

plt.figure()
for k in population_beijing.keys()[:10]:
    tmp = np.array(outflow_beijing[k])[:336]
    plt.plot(tmp/np.max(tmp))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
file = 'checkin_April'       #check-in 数据，有时间、类别、poi及类型
checkin = np.zeros([675,1488])
with open(file) as f:
    for line in f:
        tim,region,cata = line.split('|')
        r = int(region.split('-')[1])
        checkin[r][int(tim)] += 1
plt.figure()
for k in range(10):
    tmp = checkin[k,:336]
    plt.plot(tmp/np.max(tmp))
plt.show()
```
